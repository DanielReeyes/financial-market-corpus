```python
import spacy
import pandas as pd
import itertools
import os
import numpy as np
nlp = spacy.load('pt')
```

```python
pwd = %pwd
pwd = os.path.join(os.path.dirname(pwd), 'named-entity-extractor\input')
print(pwd)
data = pd.read_csv(os.path.join(pwd, 'new_data_twitter.csv'), sep=';', encoding='iso8859-1').replace({np.NaN: None})

#data = data.head(10)
```

```python
data.shape
```

```python
list_sentence = data.Sentença.tolist()
d = []

for sentence in list_sentence:
    count = 0
    doc = nlp(sentence)
    entities = doc.ents
    vector_entities = [entidade.text for entidade in entities if len(entidade.text) > 1]
    vector_type_entities = [entidade.label_ for entidade in entities if len(entidade.label_) > 1]
    dictEntities = {vector_entities[i]: vector_type_entities[i] for i in range(len(vector_entities))}
    #print ("Resultant dictionary is : " +  str(dictEntities))
    
    if (len(set(vector_entities))) > 1:
        for comb in itertools.combinations(set(vector_entities), 2):
            ent1 = comb[0]
            ent2 = comb[1]

            d.append(
                {
                    'Sentença': sentence,
                    'Entidade 1': ent1,
                    'Tipo Entidade 1': dictEntities[ent1],
                    'Relacao': "",
                    'Entidade 2 ': ent2,
                    'Tipo Entidade 2 ': dictEntities[ent2]
                }
            )
    else:
        print('Sentença \' ' + sentence + ' \' possui '+ str(len(set(vector_entities))) +' entidade identificada. Verifique.')
        print(set(vector_entities))
```

```python
df = pd.DataFrame(d)
df.head(5)
```
